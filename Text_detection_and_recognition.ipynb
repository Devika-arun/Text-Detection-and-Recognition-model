Displaying the image

```python
import PIL
from PIL import ImageDraw
im = PIL.Image.open(picture)
im
```

performing OCR on the image using easyocr

```python
reader = easyocr.Reader(['en'])
```

Drawing the bounding boxes on the image

```python
bounds = reader.readtext(picture)

def draw_boxes(image, bounds, color='blue', width=2):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        p0, p1, p2, p3 = bound[0]
        print(p0, p1, p2, p3)
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
    return image
draw_boxes(im, bounds)
```

```python
# Extract recognized text as a string
recognized_text = ''
for detection in bounds:
    recognized_text += detection[1] + ' '

print("Recognized Text:", recognized_text)
```
